```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
N = 50
noise1 = np.random.normal(0,0.15,N)
noise2 = np.random.normal(0,4,N)
noise3 = np.random.normal(0,3,N)
x = np.array(list(range(N))) + noise1
# y = np.sin(x/np.pi) + noise2
y = 2*x + noise2
z = 1.5*x + noise3

ax = plt.figure().add_subplot(projection='3d')
ax.scatter(x, y, z, zdir='y', label='points in (x, z)')

ax.legend()
ax.set_xlim(int(np.min(x, axis=0)) - 1, int(np.max(x, axis=0)) + 1)
ax.set_ylim(int(np.min(y, axis=0)) - 1, int(np.max(y, axis=0)) + 1)
ax.set_zlim(int(np.min(z, axis=0)) - 1, int(np.max(z, axis=0)) + 1)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

ax.view_init(elev=20., azim=-35, roll=0)

plt.show()
```
